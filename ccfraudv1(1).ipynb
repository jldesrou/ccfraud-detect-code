```python
import pandas as pd
import matplotlib.pyplot as plt
filename = "/kaggle/input/credit-card-fraud-detection/creditcard.csv"
df = pd.read_csv(filename)
df.head()
```

```python
df.drop(columns=['Time'], inplace=True)
```

```python
df.drop(df.loc[:, 'V5':'V28'].columns, axis = 1)
```

```python
classification=df.Class.value_counts(sort= True)
classification
```

```python
X = df.drop('Class',axis = 1)
y = df['Class']
```

```python
from imblearn.over_sampling import RandomOverSampler
os =  RandomOverSampler()
X_res, y_res = os.fit_resample(X, y)
```

```python
oversampling=y_res.value_counts()
oversampling
```

```python
from sklearn.model_selection import train_test_split
features = X_res
labels = y_res

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.4, random_state=42)
X_test, X_val, y_test, y_val = train_test_split(X, y, test_size=0.5, random_state=42)
```

```python
print(len(labels), len(y_train), len(y_test), len(y_val))
```

```python
X_train.to_csv('../train_features.csv', index=False)
X_val.to_csv('../val_features.csv', index=False)
X_test.to_csv('../test_features.csv', index=False)

y_train.to_csv('../train_labels.csv', index=False)
y_val.to_csv('../val_labels.csv', index=False)
y_test.to_csv('../test_labels.csv', index=False)
```

```python
import joblib
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

tr_features = pd.read_csv('../train_features.csv')
tr_labels = pd.read_csv('../train_labels.csv', header=0)
print(len(tr_features),len(tr_labels))
```

```python
def print_results(results):
    print('BEST PARAMS: {}\n'.format(results.best_params_))

    means = results.cv_results_['mean_test_score']
    stds = results.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, results.cv_results_['params']):
        print('{} (+/-{}) for {}'.format(round(mean, 3), round(std * 2, 3), params))
```

```python
lr = LogisticRegression()
parameters = {
    'C': [0.01, 0.1, 1, 10]
}
cv = GridSearchCV(lr, parameters, cv=5)
```

```python
cv.fit(tr_features, tr_labels)

print_results(cv)
```

```python
cv.best_estimator_
```

```python
joblib.dump(cv.best_estimator_, '../LR_model.pkl')
```

```python
import joblib
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

tr_features = pd.read_csv('../train_features.csv')
tr_labels = pd.read_csv('../train_labels.csv', header=0)
```

```python
def print_results(results):
    print('BEST PARAMS: {}\n'.format(results.best_params_))

    means = results.cv_results_['mean_test_score']
    stds = results.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, results.cv_results_['params']):
        print('{} (+/-{}) for {}'.format(round(mean, 3), round(std * 2, 3), params))
```

```python
rf = RandomForestClassifier()
parameters = {
    'n_estimators': [5, 10, 15],
    'max_depth': [1, 5, 9, None]
}
cv = GridSearchCV(rf, parameters, cv=5)
cv.fit(tr_features, tr_labels)

print_results(cv)
```

```python
cv.best_estimator_
```

```python
joblib.dump(cv.best_estimator_, '../RF_model.pkl')
```

```python
import joblib
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score
from time import time

val_features = pd.read_csv('../val_features.csv')
val_labels = pd.read_csv('../val_labels.csv', header=0)

te_features = pd.read_csv('../test_features.csv')
te_labels = pd.read_csv('../test_labels.csv', header=0)
```

```python
print(len(val_features), len(val_labels), len(te_features), len(te_labels))
```

```python
models = {}

for mdl in ['LR','RF']:
    models[mdl] = joblib.load('../{}_model.pkl'.format(mdl))
```

```python
def evaluate_model(name, model, features, labels):
    start = time()
    pred = model.predict(features)
    end = time()
    accuracy = round(accuracy_score(labels, pred), 3)
    precision = round(precision_score(labels, pred), 3)
    recall = round(recall_score(labels, pred), 3)
    print('{} -- Accuracy: {} / Precision: {} / Recall: {} / Latency: {}ms'.format(name,
                                                                                   accuracy,
                                                                                   precision,
                                                                                   recall,
                                                                                   round((end - start)*1000, 1)))
```

```python
models
```

```python
for name, mdl in models.items():
    evaluate_model(name, mdl, val_features, val_labels)
```

```python
evaluate_model('Random Forest', models['RF'], te_features, te_labels)
```
